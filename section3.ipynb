```python
import numpy as np
import matplotlib.pyplot as plt
from statsmodels.tsa.stattools import acf, acovf
```

\#2

```python
np.random.seed(12)

n = 200
Z = np.random.normal(0,1,n)
t = np.arange(n)
b = [0.7,-0.6,0]

X = np.zeros([3,n])

#b1
for index, value in enumerate(b):
    X[index,] = value*t + np.cumsum(Z)
    
```

```python
plt.plot(X[0,], marker = 'o', color = 'blue', label = 'Random Walk, b = 0.7')
plt.plot(X[1,], marker = 'o', color = 'red', label = 'Random Walk, b = -0.6')
plt.plot(X[2,], marker = 'o', color = 'yellow', label = 'Random Walk, b = 0')
# plotting slashed lines y = 0.8t and y = -0.2t
plt.xlabel('t')
plt.ylabel('X_t')
plt.title('Random Walk for t = 1, 2, ..., 100')
plt.legend(loc='upper left', fontsize='small')
plt.show()
```

```python
# acf
acf_values_b0 = acf(X[0,], nlags=20)
acf_values_b1 = acf(X[1,], nlags=20)
acf_values_b2 = acf(X[2,], nlags=20)


plt.stem(acf_values_b0, linefmt = 'r--')
plt.stem(acf_values_b1, linefmt = 'b--')
plt.stem(acf_values_b2, linefmt = 'g--')
plt.title("ACF of Random Walk")
plt.xlabel("h")
plt.ylabel("ACF")
plt.show()
```
